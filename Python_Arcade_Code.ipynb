**The code for the Main Menu**

```python
# imported to use for the user interphase and to clear and display output
from ipywidgets import Button, VBox, Output, Label
from IPython.display import display, clear_output

# Main menu function
def main_menu():
    clear_output()
# Create buttons for each game
    math_quiz_button = Button(description="Math Quiz", button_style="primary")
    rps_button = Button(description="Rock-Paper-Scissors", button_style="info")
    guess_number_button = Button(description="Guess the Number", button_style="primary")
    exit_button = Button(description="Exit", button_style="danger")

# if the user click exit, this happens
    def on_exit_click(b):
        clear_output()
        print("Goodbye! Hope to see you again another time...")

# Link buttons to the different games, so when clicked they run the different games
    math_quiz_button.on_click(play_mathQuiz)
    rps_button.on_click(play_RPS)
    guess_number_button.on_click(play_guessNumber)
    exit_button.on_click(on_exit_click)

# Display the menu
    display(VBox([
        Label("Welcome to Python Arcade! Choose a game:"),
        math_quiz_button,
        rps_button,
        guess_number_button,
        exit_button
    ]))
```

**The code for the Guess the number game**


```python
# imported to use for the user interphase and to clear and display output
from ipywidgets import Button, VBox, Output, Text, Label
from IPython.display import clear_output, display
import random

# the function for the Guess the Number game
def play_guessNumber(b=None):
    clear_output()
# Generates a random number as assigns it to the varible secret_number
    secret_number = random.randint(1, 100)
# Creates the varible attempts, which will track the number of attemps used
    attempts = 0
# Create the buttons and text for the game
    label = Label("Guess the number I'm thinking of (1-100):")
    guess_input = Text(placeholder='')
    submit_button = Button(description='Submit')
    output = Output()

# the functions that checks the guess and add an atempt for each guess
    def check_guess(btn):
# using nonlocal to access the attemps variable outside of the function
        nonlocal attempts
# adds an attempt to each time the function is called
        attempts += 1
# the try block changes the user input into an integer, and checks if the
# different condistion are met, using different if statements
        try:
            guess = int(guess_input.value)
            if guess < secret_number:
                with output:
                    clear_output(wait=True)
                    print("Too low! Try again.")
            elif guess > secret_number:
                with output:
                    clear_output(wait=True)
                    print("Too high! Try again.")
            else:
                with output:
                    clear_output(wait=True)
                    print(f"Congratulations! You guessed it in {attempts} attempts!")
# The user gets a error message if they do nbot enter a valid number
        except ValueError:
            with output:
                clear_output(wait=True)
                print("Please enter a valid number.")
# Clear the input box
        guess_input.value = ''

# Link button to the function check_guess
    submit_button.on_click(check_guess)

# Display the game
    display(VBox([label, guess_input, submit_button, output]))
```

**The code for the Math Quiz**

```python
# import random - for the random numbers for the questions
import random

# import to use with the quiz interface
from ipywidgets import Text, Button, Output, VBox, Label

# used to show and clear the outputs the are created
# used in the beginning or ending of a game/round
from IPython.display import display, clear_output

# Start by defining a class
class MathQuiz:

# this function sets up the class
    def __init__(self, total_games=5):
# Tracks the players score - everyone starts at 0
        self.score = 0
# Total number of games, default is 55
        self.total_games = total_games
# Tracks the current game
        self.current_game = 1
# Creates a widget to put messages inside for the user
        self.output = Output()
# Creates a box for the user to type into
        self.input_box = Text(placeholder="")
# Creates a button for submission
        self.submit_button = Button(description="Submit", button_style="primary")
# Creates a widget to put feedback to the user
        self.feedback_label = Label()
# creates a container for all widgets
        self.game_widgets = VBox()

# Creates random numbers (1-10) and a random operator (+ - *)
    def generate_question(self):
        num1 = random.randint(1, 10)
        num2 = random.randint(1, 10)
        operator = random.choice(["+", "-", "*"])
# puts the generated numbers and operators together to form the question
        question = f"\n{num1} {operator} {num2}"
# calculates the answer (so we can use this to compare to the users input)
        correct_answer = eval(question)
        return question, correct_answer

# Starts the game
    def play_game(self):
# uses the clear_output from IPython.display to clean the notebook
        clear_output()
        self.score = 0
        self.current_game = 1
        self.next_question()

# Generates the questions
    def next_question(self):
        # runs the code as long as the number of total games are not reached
        if self.current_game <= self.total_games:
# generates a new question
            self.question, self.correct_answer = self.generate_question()
# makes the user able to see number of games played
            self.feedback_label.value = f"Game {self.current_game} of {self.total_games}: {self.question}"
# makes the value of the user input empty (clears the past answer)
            self.input_box.value = ""
# enables the input and submit bottons again
            self.input_box.disabled = False
            self.submit_button.disabled = False
# if total of games are reached, then the game is done
        else:
            self.end_game()

# validated the users input and checks if their answer is correct
    def check_answer(self, b):
        try:
            # converts the input to an integer
            user_answer = int(self.input_box.value)
            # give a + 1 for their score if answer is correct
            if user_answer == self.correct_answer:
                self.score += 1
                # clears the output area and prints success message
                with self.output:
                    clear_output(wait=True)
                    print("You were right! One point for you")
            # clears the output area and prints error message if
            # the answer was not correct
            else:
                with self.output:
                    clear_output(wait=True)
                    print(f"Wrong! The correct answer was {self.correct_answer}.")
# clears area and invalid input error message.
        except ValueError:
            with self.output:
                clear_output(wait=True)
                print("Please enter a valid number ")

# adds + 1 to the current number of games and creates a new question
        self.current_game += 1
        self.next_question()

# function that runs when game is over
    def end_game(self):
        self.input_box.disabled = True
        self.submit_button.disabled = True
# the user feedback
        feedback = f"\nGame Over! Your final score is: {self.score} out of {self.total_games}\n"
# different feedback based on different scores
        if self.score == self.total_games:
            feedback += "You are doing amazing!"
        elif self.score >= 4:
            feedback += "Good job, maybe try again and see if you can get 5/5!"
        elif self.score >= 3:
            feedback += "Ahh good job, but now goode enough! Try again."
# any score below 3 gets this feedback:
        else:
            feedback += "Practice makes perfect, feel free to try again!"

        with self.output:
            clear_output(wait=True)
            print(feedback)

# the function that connects the button to check_answer and display outputs
    def display_game(self):
        self.submit_button.on_click(self.check_answer)
        self.game_widgets.children = [self.feedback_label, self.input_box, self.submit_button, self.output]
        display(self.game_widgets)

# Function to integrate with the main menu
def play_mathQuiz(b=None):
    clear_output()
    game = MathQuiz()
    game.play_game()
    game.display_game()
```

**The code for the Rock, paper, scissors game**

```python
import random
from ipywidgets import Button, VBox, Output, Label
from IPython.display import display, clear_output

# Function to play Rock-Paper-Scissors
def play_RPS(b=None):
    clear_output()

    output = Output()  # For displaying messages

    # Buttons for the user's choices
    rock_button = Button(description="Rock", button_style="info")
    paper_button = Button(description="Paper", button_style="warning")
    scissors_button = Button(description="Scissors", button_style="danger")

    # Display widgets
    label = Label("Choose Rock, Paper, or Scissors to play against the Helle!")
    display(VBox([label, rock_button, paper_button, scissors_button, output]))

    # The computer randomly chooses
    def computer_choice():
        return random.choice(["Rock", "Paper", "Scissors"])

    # Function to handle the outcome
    def determine_winner(user_choice):
        comp_choice = computer_choice()
        with output:
            clear_output(wait=True)
            print(f"You choice: {user_choice}")
            print(f"Helle's choice: {comp_choice}")

            if user_choice == comp_choice:
                print("It's a tie, nobody won /: ")
            elif (
                (user_choice == "Rock" and comp_choice == "Scissors") or
                (user_choice == "Paper" and comp_choice == "Rock") or
                (user_choice == "Scissors" and comp_choice == "Paper")
            ):
                print("You win!")
            else:
                print("You lost...")

    # Link buttons to user choices
    rock_button.on_click(lambda b: determine_winner("Rock"))
    paper_button.on_click(lambda b: determine_winner("Paper"))
    scissors_button.on_click(lambda b: determine_winner("Scissors"))
```
